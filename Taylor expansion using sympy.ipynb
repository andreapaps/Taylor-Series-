```python
import sympy as smp
from sympy import * 

#x= smp.symbols('x')
#f = x**2 + 2*x+5        this is to show the syntax 
#g= smp.diff(f)
#g.subs(x,4)
#f
def fact(n):
    for i in range(0,n):
        if i != 0:
            n = n*i
            
    if n == 0:
        n = n + 1
    return n

x= smp.symbols('x')
f = smp.exp(x)      # choose your function
g= f.subs(x,0)
count =100          #number of terms of the taylor expansion

for i in range(0,count):
    h= smp.diff(f)
    g = g + (h.subs(x,0)*x**(i+1))/fact(i+1)
    f = h
   
g
    
```
